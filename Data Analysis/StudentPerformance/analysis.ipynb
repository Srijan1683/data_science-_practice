Now i am plotting a box-plot using libraries like matplotlib and seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns

order = ["Low", "Medium", "High"]

df["Parental_Involvement"] = df["Parental_Involvement"].astype(str).str.strip().str.title()
df["Access_to_Resources"] = df["Access_to_Resources"].astype(str).str.strip().str.title()

plt.figure(figsize=(11, 6))
sns.set_style("whitegrid")

sns.boxplot(
    data=df,
    x="Parental_Involvement",
    y="Exam_Score",
    hue="Access_to_Resources",
    order=order,
    hue_order=order,
    palette="Set2",
    width=0.65
)

plt.title("Exam Score by Parental Involvement and Access to Resources", fontsize=13)
plt.xlabel("Parental Involvement")
plt.ylabel("Exam Score")
plt.legend(title="Access to Resources", bbox_to_anchor=(1.02, 1), loc="upper left")
plt.tight_layout()
plt.show()
```

As many outliers can be seen... i am cleaning them using IQR (Inter-Quartile Range)

```python
# Removing outliers from Exam_Score using IQR
q1 = df["Exam_Score"].quantile(0.25)
q3 = df["Exam_Score"].quantile(0.75)
iqr = q3 - q1

lower = q1 - 1.5 * iqr
upper = q3 + 1.5 * iqr

df_clean = df[(df["Exam_Score"] >= lower) & (df["Exam_Score"] <= upper)].copy()


#Plotting the graph (cleaner version)
import matplotlib.pyplot as plt
import seaborn as sns

order = ["Low", "Medium", "High"]

plt.figure(figsize=(11, 6))
sns.set_style("whitegrid")

sns.boxplot(
    data=df_clean,
    x="Parental_Involvement",
    y="Exam_Score",
    hue="Access_to_Resources",
    order=order,
    hue_order=order,
    palette="Set2",
    width=0.65
)

plt.title("Exam Score (Outliers Removed)")
plt.xlabel("Parental Involvement")
plt.ylabel("Exam Score")
plt.legend(title="Access to Resources", bbox_to_anchor=(1.02, 1), loc="upper left")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

df100 = df.head(100).copy()
df100["Improvement"] = df100["Exam_Score"] - df100["Previous_Scores"]

plt.figure(figsize=(10, 7))
sns.scatterplot(
    data=df100,
    x="Hours_Studied",
    y="Tutoring_Sessions",
    size=df100["Improvement"].abs(),
    hue="Improvement",
    palette="RdYlGn",
    sizes=(30, 300),
    alpha=0.75
)

plt.title("Score Improvement by Hours Studied and Tutoring Sessions")
plt.xlabel("Hours Studied")
plt.ylabel("Tutoring Sessions")
plt.legend(bbox_to_anchor=(1.02, 1), loc="upper left")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df_first_100 = df.head(100).copy()

bar_order = ["Low", "Medium", "High"]

bar_summary = (
    df_first_100.groupby("Parental_Involvement", as_index=False)["Exam_Score"]
    .mean()
)

bar_summary["Parental_Involvement"] = pd.Categorical(
    bar_summary["Parental_Involvement"],
    categories=bar_order,
    ordered=True
)

bar_summary = bar_summary.sort_values("Parental_Involvement")

plt.figure(figsize=(8, 5))
sns.set_style("whitegrid")

ax = sns.barplot(
    data=bar_summary,
    x="Parental_Involvement",
    y="Exam_Score",
    palette="Blues_d"
)

ax.set_title("Average Exam Score by Parental Involvement for the First 100 Entries")
ax.set_xlabel("Parental Involvement")
ax.set_ylabel("Average Exam Score")

for container in ax.containers:
    ax.bar_label(container, fmt="%.1f", padding=3)

plt.tight_layout()
plt.show()
```
